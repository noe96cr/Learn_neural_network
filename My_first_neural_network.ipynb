```python
import tensorflow as tf
import numpy as np
```

```python
celsius = np.array([-40,-10,0,8,15,22,38], dtype=float)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
layer = tf.keras.layers.Dense(units=1, input_shape=[1])
#There's more kinds of models, this time we'll use Sequential
model = tf.keras.Sequential([layer])
```

```python
model.compile(
    #Now we're going to use ADAM optimizer in 0.1 which means they learning with steps of 0.1 usually if you select higher values
    # the correction will be higher and probably skip the right values, there are more than ADAM and all of them help the proccess to learn.
    optimizer=tf.keras.optimizers.Adam(0.1),
    # Mean squared error: A lot of litte errors are better than a little amount of big erros
    loss= 'mean_squared_error'
)
```

```python
print('Training started...')
#Epoch it's the laps of the train...
#Verbose is if you want to print or now every lap..
record = model.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print('Model trained!')
```

```python
#With this line graph we can see how fast it machine has learn. It looks that around 500
#It stopped learning, it was enough at 500
import matplotlib.pyplot as plt
plt.xlabel('# Laps')
plt.ylabel('Loss magnitude')
plt.plot(record.history['loss'])
```

```python
#Now let's check it. First prediction with our simple neural network
print('Lets pedict!')
result = model.predict([100.0])
print('The result is ' + str(result) + 'fahrenheit')
```

```python
#Now we're going to see the weight of the variables that the model asigned
print('Intern variables of the model')
print(layer.get_weights())
#Remember about y = mx + b
#It says the first was 1.79, it is 'x', and the second one 31.91, it is 'b'
```

```python
#Now, we're going to add more layers and neurons
hidden1 = tf.keras.layers.Dense(units=3, input_shape=[1])
hidden2 = tf.keras.layers.Dense(units=3)
exit = tf.keras.layers.Dense(units=1)
model2 = tf.keras.Sequential([hidden1, hidden2, exit])
```

```python
model2.compile(
    #Now we're going to use ADAM optimizer in 0.1 which means they learning with steps of 0.1 usually if you select higher values
    # the correction will be higher and probably skip the right values, there are more than ADAM and all of them help the proccess to learn.
    optimizer =tf.keras.optimizers.Adam(0.1),
    # Mean squared error: A lot of litte errors are better than a little amount of big erros
    loss= 'mean_squared_error'
)
```

```python
print('Comenzando entrenamiento...')
#Epoch es la cantidad de entrenamientos o vueltas.
#Verbose es si imprime lo que pasa en cada vuelta
record2 = model2.fit(celsius, fahrenheit, epochs=1000, verbose=False)
print('Modelo entrenado!')
```

```python
plt.xlabel('# Epoca')
plt.ylabel('Magnitud de perdida')
plt.plot(record2.history['loss'])
```

```python
#Prediccion probando nuestra red neuronal de 3 capas
print('Hagamos una prediccion!')
result2 = model2.predict([-0.02])
print('El resultado es ' + str(result2) + 'fahrenheit')
```

```python
#Now we're going to see the weight of the variables that the model asigned
print('Intern variables of the model')
print(hidden1.get_weights())
print(hidden2.get_weights())
print(exit.get_weights())

#Now, it gets more complicated than just 'y = mx + b', that's why when we talk
#about neuronal red, it can be huge, but the important is that the machine is doing
#it's work..
```
